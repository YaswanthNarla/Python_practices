```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style='whitegrid')
```

```python
sns.countplot(data=df, x='Attrition', palette='Set2')
plt.title("Attrition Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Department', hue='Attrition', palette='Set1')
plt.title("Attrition by Department")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df, x='Gender', palette='cool')
plt.title("Gender Distribution")
plt.show()
```

```python
df.groupby('Department')['Salary'].mean().plot(kind='bar', color='skyblue')
plt.title("Average Salary by Department")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df, x='PerformanceScore', hue='Attrition', palette='magma')
plt.title("Attrition by Performance Score")
plt.show()
```

```python
numeric_df = df[['Age', 'Salary', 'TrainingHours',]]
corr = numeric_df.corr()

sns.heatmap(corr, annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```
